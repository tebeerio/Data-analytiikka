```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

plt.style.use('seaborn-whitegrid')
```

```python
df=pd.read_excel('http://taanila.fi/CO2.xlsx')
df.index = pd.to_datetime(df['Kuukausi'],format="%Y-%m")
df=df.drop('Kuukausi',axis=1)
df.head()
```

```python
df['CO2'].plot()
```

```python
df['CO2'].resample('Y').mean().plot()
```

```python
df['CO2']['2019':].plot()
```

```python
df['CO2'].plot()

#liukuvia keskiarvoja
df['CO2'].rolling(20).mean().plot()
df['CO2'].rolling(50).mean().plot()
```

```python
df['CO2%'] = df['CO2'].pct_change()
df.head()
```

```python
df['CO2%']['2019':].plot()
```

```python
df['CO2%'].describe()
```

```python
df[(abs(df['CO2%'])>0.005)]
```

```python
(df['CO2%'].rolling(12).std()*(12**0.5)).plot(label='CO2', legend=True)
```
