```python
import numpy as np
import pandas as pd # Tuodaan pandas- ja numpy-kirjasto.
import matplotlib.pyplot as plt # Tuodaan kuvioiden ja kaavioiden peruskirjasto.
# Tällä komennolla kaaviot saadaan tulostumaan automaattisesti ilman erillistä komentoa.
%matplotlib inline 

plt.style.use('seaborn-whitegrid') # Kaavioiden ulkoasun tyyli.
```

```python
df = pd.read_csv('http://taanila.fi/sales_data_sample.csv', encoding = 'cp1252') # Tuodaan csv-tiedosto ja avataan viisi ensimmäistä riviä.
df.head()
```

```python
# Tehdään taulukko, josta näkyy myyntien luokat uutena muuttujana salesclass sekä yksittäisten myyntien lukumäärä ja prosentit.
bins = [1000, 2000, 3000, 4000, 7000, 10000, 15000,] # Tehdään luokkarajat.

df['SALESCLASS'] = pd.cut(df['SALES'], bins = bins) # Uusi muuttuja SALESCLASS.

df1 = pd.crosstab(df['SALESCLASS'], 'count') # Luokkiin kuuluvien myyntien lukumäärä.

df1.columns.name = '' # Poistetaan turha otsikko.

n = df1['count'].sum() # Lisätään prosentit.
df1['%'] = df1/n*100

df1.index = df1.index.astype(str) # Lisätään yhteensä-rivi.
df1.loc['Yhteensä'] = df1.sum()

df1.style.format({'%': '{:.1f} %'}) # Viimeistellään prosenttilukujen ulkoasu.
```

```python
# Tehdään myynneistä pylväskuvio luokittain.
ax2 = df['SALES'].hist(bins=bins)
ax2.set_xlabel('sales')
ax2.set_ylabel('count')
```
