```python
import pandas as pd
df = pd.read_excel('http://taanila.fi/data1.xlsx')
df.head()
```

```python
df.count()
```

```python
df.isnull().sum()
```

```python
for var in df: #var vain joku keksitty nimi
    print(var, pd.unique(df[var]))
```

```python
df['sukup_teksti'] = df['sukup'].replace({1:'Mies', 2:'Nainen'})
df.head(6) #tehdään uusi sarake sukup_teksti ja otetaan sukup-sarakkeesta tiedot 1 ja 0
```

```python
bins = [18,28,38,48,58,68] #tehdään luokittelu ja uusi sarake
df['ikäluokka'] = pd.cut(df['ikä'], bins = bins).astype(str)#str on merkkijono ja astype on funktio jolla määritellään jokin tietynlaiseksi
df.head()
```

```python
df['tyytyväisyys'] = df[['johto', 'työtov', 'työymp', 'palkkat', 'työteht']].mean(axis=1) #uusi muuttuja tyytyväisyys ja keskiarvo mainituista muuttujista
df[['sukup', 'ikä', 'tyytyväisyys']].head()  #axis=1 tarkoittaa että luetaan vasemmalta oikealle
```

```python
df['käyttö'] = df[['työterv', 'lomaosa', 'kuntosa', 'hieroja']].count(axis=1)
df.head(6)
```

```python
#tehdään poimintoja ja suodatuksia
df[4:6] #nelonen jo iteasiassa viides rivi ja 'kuudes' ei kuulu mukaan koska pythonin indeksointi alkaa nolasta
```

```python
df[df['palkka']>4000]
```

```python
#miehet ota ovat tyytymättömiä palkkaansa, kaksi muuttujaa sukup ja palkkat
df[(df['sukup']==1) & (df['palkkat']<3)] #kaksi =merkkiä koska yksi on varattu sijoittamiselle
```

```python
#ne jotka eivät ole käyttäneet mitään etuisuuksia
df[df['käyttö']==0]
```

```python
#kokonaistyytyväisyys niillä jotka ovat käyttäneet vähintään kolmea etuisuutta, kahdet hakasulkeet koska 1. viitataan ja 2. on lista
df[['tyytyväisyys', 'käyttö']][df['käyttö']>=3] #alkuosa kertoo mitkä muuttujat tulostetaan näkyviin ja loppuosa millä ehdolla
```

```python
#sorttaus palkka ja palkkatyytyväisyys palkan mukaan ja kymmenen ensimmäistä riviä suurimmasta pienimpään
df[['palkka', 'palkkat']].sort_values(by='palkka', ascending=False).head(10)
```

```python
#aineistosta karsitaan muuttujia pois
df.drop(['nro', 'ikä', 'palveluv', 'palkka'], axis=1).head()
```

```python
#miten muutettu aineisto laitetaan exceliin
writer = pd.ExcelWriter('valmisteltu.xlsx', engine='xlsxwriter')
df.to_excel(writer)
writer.save()
```
